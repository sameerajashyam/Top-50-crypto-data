```python
import requests
import pandas as pd
import time
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt

from google.colab import files
```

```python

API_URL = "https://api.coingecko.com/api/v3/coins/markets"
```

```python

PARAMS = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 50,
    "page": 1,
    "sparkline": "false"
}
FILE_NAME = "crypto_data.xlsx"
REFRESH_INTERVAL = 300  # Fetching data for every 5 mins
```

**Fetches cryptocurrency market data from CoinGecko API**

```python

def fetch_crypto_data():

    response = requests.get(API_URL, params=PARAMS)
    if response.status_code == 200:
        df = pd.DataFrame(response.json())
        df["timestamp"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        return df
    print("Error fetching data:", response.status_code)
    return pd.DataFrame()

```

**Performs basic analysis on cryptocurrency data.**

```python
def analyze_data(df):

    df = df[["name", "symbol", "current_price", "market_cap", "total_volume", "price_change_percentage_24h", "timestamp"]]
    analysis = {
        "Top 5 by Market Cap": df.nlargest(5, "market_cap"),
        "Average Price": df["current_price"].mean(),
        "Highest 24h Change": df.loc[df["price_change_percentage_24h"].idxmax()],
        "Lowest 24h Change": df.loc[df["price_change_percentage_24h"].idxmin()]
    }
    return df, analysis
```

**Saves or updates cryptocurrency data in an Excel file.**

```python
def save_to_excel(df):

    try:
        existing_df = pd.read_excel(FILE_NAME)
        df = pd.concat([existing_df, df], ignore_index=True)
    except FileNotFoundError:
        pass
    df.to_excel(FILE_NAME, index=False)

```

**Runs the data collection and analysis in a loop.**

```python
def main():

    while True:
        df = fetch_crypto_data()
        if not df.empty:
            df, analysis = analyze_data(df)
            save_to_excel(df)
            print("Data updated successfully.")
        time.sleep(REFRESH_INTERVAL)
```

**Plots individual time-series subplots for top 50 cryptocurrencies.**

```python
def plot_time_series():

    try:
        df = pd.read_excel(FILE_NAME)
        df["timestamp"] = pd.to_datetime(df["timestamp"])
        unique_cryptos = df["name"].unique()
        num_cryptos = len(unique_cryptos)
        cols = 5
        rows = (num_cryptos // cols) + (num_cryptos % cols > 0)
        fig, axes = plt.subplots(rows, cols, figsize=(20, 15), sharex=True)
        axes = axes.flatten()
        sns.set_style("darkgrid")

        for i, crypto in enumerate(unique_cryptos):
            subset = df[df["name"] == crypto]
            axes[i].plot(subset["timestamp"], subset["current_price"], marker='o', linestyle='-')
            axes[i].set_title(crypto, fontsize=10)
            axes[i].tick_params(axis='x', rotation=45, labelsize=8)

        for j in range(i + 1, len(axes)):
            fig.delaxes(axes[j])

        plt.tight_layout()
        plt.show()
    except FileNotFoundError:
        print("No historical data found. Please wait for the file to be generated.")
```

**Appends new cryptocurrency data to an existing Excel file.**

```python
def save_to_excel(df):

    try:
        existing_df = pd.read_excel(FILE_NAME)
        df = pd.concat([existing_df, df], ignore_index=True)
    except FileNotFoundError:
        df.to_excel(FILE_NAME, index=False)
        return
    df.to_excel(FILE_NAME, index=False)
```

**Runs the data collection and analysis in a loop.**

```python
def main():

    while True:
        df = fetch_crypto_data()
        if not df.empty:
            df, analysis = analyze_data(df)
            save_to_excel(df)
            print("Data updated successfully.")
            plot_time_series()
        time.sleep(REFRESH_INTERVAL)
```

```python

if __name__ == "__main__":
    main()
files.download(FILE_NAME)
```
